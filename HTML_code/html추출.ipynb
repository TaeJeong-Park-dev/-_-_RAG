```python
from tabula import read_pdf
import pandas as pd
from markdownify import markdownify as md

# PDF에서 표 추출
def extract_tables_from_pdf(pdf_path, pages='all'):
    tables = read_pdf(pdf_path, pages=pages, lattice=True, multiple_tables=True)
    return tables

# DataFrame을 HTML로 변환
def convert_to_html(dataframe):
    return dataframe.to_html(index=False, escape=False)

# DataFrame을 Markdown으로 변환
def convert_to_markdown(dataframe):
    return dataframe.to_markdown(index=False)

# PDF에서 표를 HTML 및 Markdown으로 변환
def process_pdf_tables(pdf_path, output_html_path, output_md_path, pages='all'):
    tables = extract_tables_from_pdf(pdf_path, pages=pages)

    html_content = ""
    markdown_content = ""

    for idx, table in enumerate(tables):
        # HTML 변환
        html_content += f"<h2>Table {idx + 1}</h2>\n"
        html_content += convert_to_html(table)
        html_content += "\n"

        # Markdown 변환
        markdown_content += f"### Table {idx + 1}\n\n"
        markdown_content += convert_to_markdown(table)
        markdown_content += "\n\n"

    # HTML 저장
    with open(output_html_path, "w", encoding="utf-8") as html_file:
        html_file.write(html_content)

    # Markdown 저장
    with open(output_md_path, "w", encoding="utf-8") as md_file:
        md_file.write(markdown_content)

    print("Conversion completed!")
    print(f"HTML saved at: {output_html_path}")
    print(f"Markdown saved at: {output_md_path}")

# PDF 파일 경로
pdf_path = "/usr/workspace/raw/2021-2학기.pdf"  # 입력 PDF 파일 경로
output_html_path = "/usr/workspace/raw/output.html"  # 출력 HTML 파일 경로
output_md_path = "/usr/workspace/raw/output.md"  # 출력 Markdown 파일 경로

# PDF 처리 실행
process_pdf_tables(pdf_path, output_html_path, output_md_path)
```

```python
from tabula import read_pdf
from PyPDF2 import PdfReader
import pandas as pd
import os

# PDF에서 텍스트 추출
def extract_text_from_pdf(pdf_path, pages='all'):
    reader = PdfReader(pdf_path)
    text_content = []
    if pages == 'all':
        for page in reader.pages:
            text_content.append(page.extract_text())
    else:
        page_indices = [int(p) - 1 for p in str(pages).split(",")]
        for idx in page_indices:
            text_content.append(reader.pages[idx].extract_text())
    return text_content

# PDF에서 표 추출
def extract_tables_from_pdf(pdf_path, pages='all'):
    tables = read_pdf(pdf_path, pages=pages, lattice=True, multiple_tables=True)
    return tables

# DataFrame을 HTML로 변환
def convert_to_html(dataframe):
    return dataframe.to_html(index=False, escape=False)

# PDF에서 표와 텍스트를 HTML로 변환
def process_pdf_content(pdf_path, output_html_path, pages='all'):
    tables = extract_tables_from_pdf(pdf_path, pages=pages)
    texts = extract_text_from_pdf(pdf_path, pages=pages)

    html_content = "<html><body>\n"

    # 텍스트와 표를 순서대로 추가
    for idx, (text, table) in enumerate(zip(texts, tables)):
        # 텍스트 추가
        text = text.replace("\n", "<br>")  # 줄바꿈 처리
        html_content += f"<h2>Text Section {idx + 1}</h2>\n"
        html_content += f"<p>{text}</p>\n"

        # 표 추가
        html_content += f"<h2>Table {idx + 1}</h2>\n"
        html_content += convert_to_html(table)
        html_content += "\n"

    # 남은 텍스트 추가
    if len(texts) > len(tables):
        for idx in range(len(tables), len(texts)):
            text = texts[idx].replace("\n", "<br>")
            html_content += f"<h2>Text Section {idx + 1}</h2>\n"
            html_content += f"<p>{text}</p>\n"

    html_content += "</body></html>"

    # HTML 파일 저장
    with open(output_html_path, "w", encoding="utf-8") as html_file:
        html_file.write(html_content)

    print("변환 완료!")
    print(f"HTML 파일 위치: {output_html_path}")

# 처리할 PDF 파일 리스트
pdf_paths = [
    "/usr/workspace/raw/2020-1.pdf",
    "/usr/workspace/raw/2020-2.pdf",
    "/usr/workspace/raw/2021-1.pdf",
    "/usr/workspace/raw/2021-2.pdf",
    "/usr/workspace/raw/2022-1.pdf",
    "/usr/workspace/raw/2022-2.pdf",
    "/usr/workspace/raw/2023-1.pdf",
    "/usr/workspace/raw/2023-2.pdf",
    "/usr/workspace/raw/2024-1.pdf",
    '/usr/workspace/raw/2024-2.pdf'
]

# 각 PDF 파일에 대해 처리 실행
for pdf_path in pdf_paths:
    # PDF 파일명 추출 (확장자 제외)
    base_name = os.path.splitext(os.path.basename(pdf_path))[0]
    # HTML 출력 경로 설정
    output_html_path = f"/usr/workspace/processed/{base_name}_output_with_text.html"
    
    # PDF 처리
    process_pdf_content(pdf_path, output_html_path)
```
